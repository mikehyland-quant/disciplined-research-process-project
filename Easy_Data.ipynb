```python
import pandas as pd
import os
```

```python
symbol_list = ['BTC']
```

```python
denominator = 'USDT'

timeframe   = '1d'
start_date  = '2020-12-31'
end_date    = '2026-01-01'

directory   = r"C:\Users\micha\OneDrive\Wall Street Quants Backtesting\MyTWSQ\historical_prices"
```

def get_bars(self, symbol, freq = '1d', start_ts = None, end_ts = None):
    """
    Parameters
    ----------
    symbol : str
        Symbol to pull bars for
    freq : str {'1m', '5m','15m', '30m', '1h', '4h', '1d'}
        Frequency (size) of bars. 1m = 1 minute, 1h = 1 hour,
        1d = 1 day, etc. 
    start_ts : str, datetime or pandas timestamp, optional
        start time of bar history. defaults to earliest available.
    end_ts : str, datetime or pandas timestamp, optional
        end time of bar history. default is most recent bar

    Returns
    -------
    DataFrame 
    columns = ['open','high','low','close','volume']
    index = Bar open times
    """

```python
from twsq.data import BinancePrices
px = BinancePrices()

close_df = pd.DataFrame([], [start_date, end_date])

for symbol in symbol_list:

    asset = symbol + "/" + denominator

    df = px.get_bars(asset, 
                     timeframe, 
                     start_date, 
                     end_date)
    df = df.copy()

    df = df.rename(columns={'close': symbol})
    
    df = df[symbol]

    df.index.name = 'date'
    
    pd.set_option('display.max_rows', None)
    
    print(df)

    os.chdir(directory)
    filename = symbol + ' ' + str(start_date) + " " + str(end_date) + '.csv'
    df.to_csv(filename, index=True)
    
    print(symbol)
    
print('\n')
print('done')
```
